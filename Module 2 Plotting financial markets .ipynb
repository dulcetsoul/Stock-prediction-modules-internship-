```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df= pd.read_csv("RAYMOND.csv")
fig = plt.figure(figsize=(20,10))
df['Date'] = pd.to_datetime(df['Date'])
df["21 days"] = (((df["Close Price"]).rolling(21)).mean()).dropna()
df["34 days"] = (((df["Close Price"]).rolling(34)).mean()).dropna()
plt.plot(df['Date'], df['Close Price'],linewidth=0.5,color='red')
plt.plot(df['Date'], df['21 days'],linewidth=0.5,label = '21 DAYS', color='green')
plt.plot(df['Date'], df['34 days'],linewidth=0.5,label = '34 DAYS',  color='blue')
df['Buy'] = np.where( (df['21 days']> df['34 days']), 1, 0)
df['Sell'] = np.where( (df['21 days']< df['34 days']), 1, 0)
df['Buy_ind'] = np.where( (df['Buy'] > df['Buy'].shift(1)),1,0)
df['Sell_ind'] = np.where( (df['Sell'] > df['Sell'].shift(1)),1,0)
plt.scatter(df.loc[df['Buy_ind'] ==1 , 'Date'].values,df.loc[df['Buy_ind'] ==1, 'Close Price'].values, color='black', s=60, marker="^")
plt.scatter(df.loc[df['Sell_ind'] ==1 , 'Date'].values,df.loc[df['Sell_ind'] ==1, 'Close Price'].values, color='green', s=60, marker="v")
plt.xticks(rotation = 45)
plt.xlabel('Date')  
plt.ylabel('Close Price')  
plt.legend()
plt.show()
```
