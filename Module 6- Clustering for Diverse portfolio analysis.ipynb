```python

%matplotlib inline
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


symbol_dict = {
    'ASHOKA': 'Ashoka',
    'BAJAJELEC': 'Bajaj electricals',
    'PVR': 'PVR',
    'CENTURYPLY': 'Centuryply',
    'FORTIS': 'FORTIS',
    'GUJALKALI': 'GUJALKALI',
    'IDFC': 'IDFC',
    'SUZLON': 'SUZLON',
    'ITDC': 'ITDC',
    'JETAIRWAYS': 'JETAIRWAYS',
    'GMRINFRA': 'GMRINFRA',
    'EXIDEIND': 'EXIDEIND',
    'CASTROLIND': 'CASTROLIND',
    'CUMMINSIND': 'CUMMINSIND',
    'BERGEPAINT': 'BERGEPAINT',
    'APOLLOTYRE': 'APOLLOTYRE',
    'AMARAJABAT': 'AMARAJABAT',
    'ADANIPOWER': 'ADANIPOWER',
    'AJANTPHARM': 'AJANTPHARM',
    'DHFL': 'DHFL',
    'ADANIPORTS': 'ADANIPORTS',
    'ASIANPAINT': 'ASIANPAINT',
    'AXISBANK': 'AXISBANK',
    'BAJFINANCE': 'BAJFINANCE',
    'BPCL': 'BPCL',
    'CIPLA': 'CIPLA',
    'DRREDDY': 'DRREDDY',
    'GAIL': 'GAIL',
    'EICHERMOT': 'EICHERMOT',
    'HDFC': 'HDFC'}


symbols, names = np.array(sorted(symbol_dict.items())).T
quotes = []
for symbol in symbols:

    quotes.append(pd.read_csv(symbol+'.csv'))
for i,q in enumerate(quotes):
    quotes[i]=quotes[i][quotes[i].Series=='EQ']
    
close_prices = np.vstack([q['Close Price'] for q in quotes])

new = pd.DataFrame(close_prices.T)
new.columns = {"ADANIPORTS","ADANIPOWER","AJANTPHARM","AMARAJABAT","APOLLOTYRE","ASHOKA",
             "ASIANPAINT","AXISBANK","BAJAJELEC","BAJFINANCE","BERGEPAINT",
             "BPCL","CASTROLIND","CENTURYPLY","CIPLA","CUMMINSIND",
             "DHFL","DRREDDY","EICHERMOT","EXIDEIND","FORTIS",
            "GAIL","GMRINFRA","GUJALKALI","HDFC","IDFC",
            "ITDC","JETAIRWAYS","PVR","SUZLON"
             }
df = new.tail(252)
```

```python
returns = df.pct_change().mean()*252
returns = pd.DataFrame(returns)
returns.columns = ["Returns"] 
returns["Volatility"] = df.pct_change().std()*sqrt(252)
data = np.asarray([np.asarray(returns['Returns']),np.asarray(returns['Volatility'])]).T
X = data
distorsions = []
from sklearn.cluster import KMeans
from scipy.cluster.vq import kmeans,vq
for k in range(1, 15):
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    distorsions.append(k_means.inertia_)


fig = plt.figure(figsize=(15, 5))
plt.plot(range(1, 15), distorsions)
plt.grid(True)
plt.title('Elbow curve')
plt.show()
```

```python
from scipy.cluster.vq import kmeans,vq
import matplotlib.pyplot as plt
plt.style.use("seaborn-deep")
from pylab import plot,show
centroids,_ = kmeans(data,4)
# assign each sample to a cluster
idx,_ = vq(data,centroids)
fig = plt.figure(figsize=(22,12))
# some plotting using numpy's logical indexing
plot(data[idx==0,0],data[idx==0,1],'ob',
     data[idx==1,0],data[idx==1,1],'om',
     data[idx==2,0],data[idx==2,1],'or',
     data[idx==3,0],data[idx==3,1],'og',markersize = 10)
plot(centroids[:,0],centroids[:,1],'sk',markersize=22, marker="*")
for i,type in enumerate(symbols):
    x = returns.Returns[i]
    y = returns.Volatility[i]
    plt.text(x, y, type, fontsize=10)
    
plt.legend(["Cluster1","Cluster2","Cluster3","Cluster4"])
show()


```
