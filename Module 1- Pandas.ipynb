```python
import pandas as pd
import re
import numpy as np
import datetime
data = pd.read_csv("PVR.csv")
data.Date = data.Date.astype("datetime64[ns]")
data["PV"] = data["Average Price"] * data["No. of Trades"].sum()
data.head()
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year

d1=data.groupby(['Month','Year']).agg({
    'Month': "count",
    'PV': sum
})
d1["Vwap"] = d1["PV"]/d1["Month"]
d1.astype("int64")
```

```python
import pandas as pd
import re
import numpy as np
import datetime
data = pd.read_csv("PVR.csv")
data.Date = data.Date.astype("datetime64[ns]")
data['Day_Perc_Change'] = data["Close Price"].pct_change() * 100
data.dropna()
df = pd.DataFrame(data, columns = ["Symbol","Series","Date","Prev Close","Open Price","High Price",
                                   "Low Price","Last Price","Close Price",
                                   "Average Price","Total Traded Quantity",
                                   "Turnover","No. of Trades","Deliverable Qty",
                                   "% Dly Qt to Traded Qty","Day_Perc_Change"])
result = [] 
for value in data["Day_Perc_Change"]: 
    if value > -0.5 and value < 0.5: 
        result.append("Slight or No change") 
    elif value > 0.5 and value < 1: 
        result.append("Slight positive") 
    elif value > -0.5 and value < -1: 
            result.append("Slight negative")
    elif value > 1 and value < 3: 
        result.append("Positive") 
    elif value > -1 and value < -3:
        result.append("Negative") 
    elif value > 3 and value < 7: 
        result.append("Among top gainers") 
    elif value > -3 and value < 7: 
        result.append("Among top losers") 
    elif value > 7: 
        result.append("Bull run")
    else: 
        result.append("Bear drop") 
     
df["Trend"] = result  
print(df.dropna()) 

d2 = df.groupby('Trend')['Total Traded Quantity'].mean()
d3 = df.groupby('Trend')['Total Traded Quantity'].median() 
print(d2,d3)

df.to_csv("week2.csv", index=False) 
```
